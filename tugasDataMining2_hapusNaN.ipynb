# Data Cleaning dengan Menghapus seluruh nilai NaN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
df=pd.read_csv("C:/Users/Gilbert Ciady/Documents/Visual Studio Code Project/Semester 3/Data Mining/Tugas 2/heart_disease_uci.csv")
```

```python
nan_data = df[df.isna().any(axis=1)]
nan_data
```

```python
df.dropna(inplace = True)
print (df.to_string())
```

```python
print("Informasi Dataset: ")
df_info = pd.DataFrame({
    'Data Type': df.dtypes.values,
    'Entry': df.count().values,
    'Column': df.columns
})
df_info.index = range(1, len(df_info) + 1)
print(tabulate(df_info, headers='keys', tablefmt='outline'))
```

```python
missing_values =df.isnull().sum().reset_index()
missing_values.columns =['Column', 'Missing Values']
missing_values.index = range(1, len(missing_values) + 1)
print("Missing Value: ")
print(tabulate(missing_values, headers='keys', tablefmt='outline'))
```

```python
desc_stats = df.describe().reset_index()
desc_stats.index = range(1, len(desc_stats) + 1)
print(tabulate(desc_stats, headers='keys', tablefmt='outline'))
```
